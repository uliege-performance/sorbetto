# Show Ranking Scores in ROC

```python
from sorbetto.ranking.ranking_score import RankingScore
import matplotlib.pyplot as plt
```

```python
def showRankingScore ( score ) :
    for priorPos in [ 0.1, 0.9 ] :
        fig, ax = plt.subplots ()
        score.drawInROC ( fig, ax, priorPos=priorPos )
        plt.show ()
```

## Accuracy

```python
score = RankingScore.getAccuracy ()
showRankingScore ( score )
```

## True Positive Rate (TPR)

```python
score = RankingScore.getTruePositiveRate ()
showRankingScore ( score )
```

## F-score ($F_1$)

```python
score = RankingScore.getF ( beta=1.0 )
showRankingScore ( score )
```
