# Toy examples for correlation tiles

This jupyter notebook reproduces the nine tiles of Figure 4 in :cite:t:`Pierard2024TheTile'. 

It provides toy examples showing correlation tiles showing the rank correlations (Kendall τ ) between 9 probabilistic scores (those that belong to the ranking scores), and all ranking scores, for a uniform distribution of performances. 
The correlation values have been estimated based on 10,000 performances drawn at random.

```python
import scipy.stats
import matplotlib.pyplot as plt

from sorbetto.performance.distribution.uniform_distribution_of_two_class_classification_performances import UniformDistributionOfTwoClassClassificationPerformances
from sorbetto.parameterization.parameterization_default import ParameterizationDefault
from sorbetto.flavor.correlation_flavor import CorrelationFlavor
from sorbetto.tile.correlation_tile import CorrelationTile
from sorbetto.ranking.ranking_score import RankingScore
from sorbetto.core.importance import Importance

correlation_fct = 'pearsonr'

# Draw 10,000 performances at random
number_of_performances_drawn_at_random = 10000
performances_drawn_at_random = UniformDistributionOfTwoClassClassificationPerformances("Uniform distribution of performances").drawAtRandom(number_of_performances_drawn_at_random)
```

## Negative Predictive Value $NPV$

```python
npv_score = RankingScore.getNegativePredictiveValue()
correlation_flavor = CorrelationFlavor(
    performances=performances_drawn_at_random,
    correlation_coefficient = correlation_fct, 
    score = npv_score,
    )
npv_correlation_tile = CorrelationTile(name = "NPV Correlation Tile", 
                                       parameterization = ParameterizationDefault(), 
                                       flavor = correlation_flavor, 
                                       resolution = 101)

npv_correlation_tile.draw()
plt.show()
```

It is also possible to use very similar piece of code to obtain the figure 7 of the same paper.

```python
from sorbetto.core.importance import Importance
from sorbetto.performance.distribution.uniform_distribution_of_two_class_classification_performances_for_fixed_class_priors import UniformDistributionOfTwoClassClassificationPerformancesForFixedClassPriors
```

```python
performances_drawn_at_random = UniformDistributionOfTwoClassClassificationPerformancesForFixedClassPriors(priorPos=0.3).drawAtRandom(10000)
```

```python

def balanced_accuracy_score(performance: Importance):

    tpr = RankingScore.getTruePositiveRate()
    tnr = RankingScore.getTrueNegativeRate()

    return (tpr(performance) + tnr(performance) )/ 2

correlation_flavor = CorrelationFlavor(
    performances=performances_drawn_at_random,
    correlation_coefficient = correlation_fct, 
    score = balanced_accuracy_score,
    )

balanced_accuracy_tile = CorrelationTile(name = "Balanced Accuracy Tile", 
                                       parameterization = ParameterizationDefault(), 
                                       flavor = correlation_flavor, 
                                       resolution = 101)

balanced_accuracy_tile.draw()
plt.colorbar()
plt.show()
```

```python
def cohen_kappa(importance: Importance):

    tp = importance.ptp
    tn = importance.ptn
    fp = importance.pfp
    fn = importance.pfn

    num = 2*(tp*tn - fp*fn)
    den = (tp + fn)*(tn + fn) + (tp + fp)*(fp + tn)
    return num / den

correlation_flavor = CorrelationFlavor(
    performances=performances_drawn_at_random,
    correlation_coefficient = correlation_fct, 
    score = cohen_kappa,
    )

cohen_kappa_tile = CorrelationTile(name = "Cohen Kappa Tile", 
                                       parameterization = ParameterizationDefault(), 
                                       flavor = correlation_flavor, 
                                       resolution = 101)

data = cohen_kappa_tile.mat_value
fig, ax = plt.subplots()
im = ax.imshow(data, cmap = 'gist_rainbow', origin='lower')
im.set_clim(0.5, 1.0)
cbar = fig.colorbar(im, ax=ax)
cbar.set_ticks([0.5, 0.6, 0.7, 0.8, 0.9, 1])
plt.show()
```
